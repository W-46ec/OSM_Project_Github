# Heat Map of Amenities

```python
import sys
import numpy as np
import pandas as pd
import folium
from folium.plugins import MarkerCluster, HeatMap
```

```python
data_path = "./osm/amenities-vancouver.json.gz"
```

```python
data = pd.read_json(data_path, lines = True)
```

```python
coord_mid = [
    (data.lat.max() + data.lat.min()) / 2, 
    (data.lon.max() + data.lon.min()) / 2, 
]
```

```python
heap_map = folium.Map(
    location = coord_mid, 
    zoom_start = 10, 
)
```

```python
amenities = data.dropna()
```

```python
HeatMap(
    list(zip(amenities['lat'].values, amenities['lon'].values)), 
    min_opacity = 0.2, 
    blur = 18
).add_to(heap_map)
heap_map
```
