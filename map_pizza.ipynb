```python
import sys
import numpy as np
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
data_path = "./osm/amenities-vancouver.json.gz"
data = pd.read_json(data_path, lines = True)

food = data[data['amenity'].str.contains("restaurant|food|cafe|pub|bar|ice_cream|food_court|bbq|bistro") & ~data['amenity'].str.contains("disused")]
food = food.dropna()

# Question: Are there areas in the city with more pizza restaurants?
# Filter pizza restaurants
def filter_pizza(tags):
    return 'pizza' in tags.values()

pizza = food[food['tags'].apply(filter_pizza)]
pizza
```

```python
coord = [pizza.lat.mean(), pizza.lon.mean()]
m = folium.Map(location = coord, zoom_start = 11)
# folium.Marker().add_to(m)
```

```python
pizza.apply(lambda x: folium.Marker(location = [x['lat'], x['lon']]).add_to(m), axis = 1);
m
```
